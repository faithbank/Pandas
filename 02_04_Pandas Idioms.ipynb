## Pandas Idioms

```python
# Importing data processing libraries
import pandas as pd
import numpy as np
# Timing functionality from timeit module
import timeit
```

```python
df= pd.read_csv("/Users/FAITHFULNESS/Desktop/Python Exercises/census.csv")
df.head()
```

```python
# Method chaining: Every method on an object returns reference to that object
# Pull out the state and city names as a multiple index
(df.where(df['SUMLEV']==50)
     .dropna()
     .set_index(['STNAME','CTYNAME'])
     .rename(columns={'ESTIMATEBASE2010': 'Estimates Base 2010'}))
```

```python
# where() function is used to pass in a boolean mask, which is only true for rows with values equal to 50.
# where() does not drop missing values by default.
# Columns renamed
# Begin with a parenthesis to tell python the statement will span over multiple lines for readability.
```

```python
# A more traditional way of writing this

# first create a new dataframe from the original
# Use the overloaded indexing operator [] which drops nans

df= df[df['SUMLEV']==50]
# Update dataframe to have a new index, using inplace=True to do this in place
df.set_index(['STNAME', 'CTYNAME'], inplace=True)
# Set column names
df.rename(columns={'ESTIMATEBASE2010': 'Estimates Base 2010'})
```

```python
# Put the approach into a function and pass the function into
 #the timeit function to count the time the parameter number 
 #allows us choose the number of times to run the function.

# Write a wrapper for our first function
def first_approach():
    global df
    # Paste the code
    return(df.where(df['SUMLEV']==50)
     .dropna()
     .set_index(['STNAME','CTYNAME'])
     .rename(columns={'ESTIMATEBASE2010': 'Estimates Base 2010'}))

# Read in dataset anew
df= pd.read_csv("/Users/FAITHFULNESS/Desktop/Python Exercises/census.csv")

# Run it
timeit.timeit(first_approach, number=10)
```

```python
# Testing the second approach
# Changing the global variable inside a function will modify the variable even in a global scope
# To avoid this, create a new dataframe

def second_approach():
    global df
    new_df=df[df['SUMLEV']==50]
    new_df.set_index(['STNAME', 'CTYNAME'], inplace=True)
    new_df.rename(columns={'ESTIMATEBASE2010': 'Estimates Base 2010'})

# Read in dataset anew
df= pd.read_csv("/Users/FAITHFULNESS/Desktop/Python Exercises/census.csv")

# Run it
timeit.timeit(second_approach, number=10)
    
```

```python
# The second approach is much faster.
```

### Map

```python
# Python has a function called map, which is the basis for functional programming in the language.
# To use it, pass a function you want called, and an iterable, like a list, you want the fuunction applied to
# This results in the function being called against each item in the list, and there is a resulting list of
# of all the evaluations of that function.   

# Pandas has something similar called applymap. You provide a function which should operate on each cell of a
# dataframe, and the return cell is itself a dataframe.

#The function apply helps map across all the rows of a dataframe
```

```python
# Write a function that takes a particular row of data, find the minimum and maximum values,
# and return a new row of data.

def min_max(row):
    data=row[['POPESTIMATE2010',
             'POPESTIMATE2011',
             'POPESTIMATE2012',
             'POPESTIMATE2013',
             'POPESTIMATE2014']]
    return pd.Series({'min':np.min(data), 'max':np.max(data)})
```

```python
# Call apply on the DataFrame.
# Apply takes the function  and axis on which  to operate as paramenters.
df.apply(min_max, axis='columns').head()
```

```python
#  Revised version of the min_max function.
# Instead of returning a seperate series, we add two new columns 
# in the original dataframe to store min and max

def min_max(row):
    data = row[['POPESTIMATE2010',
             'POPESTIMATE2011',
             'POPESTIMATE2012',
             'POPESTIMATE2013',
             'POPESTIMATE2014']]
# Create new entries for max and min 
    row['max']=np.max(data)
    row['min']=np.min(data)
    return row
# Apply the function across the dataframe
df.apply(min_max, axis='columns')
```

```python
# Apply is used most often with lambda
# Chain several apply calls with lambdas to create a readable, yet succinct data
# manipulation script.

# Calculating the max of the columns using apply function
rows= ['POPESTIMATE2010','POPESTIMATE2011','POPESTIMATE2012','POPESTIMATE2013','POPESTIMATE2014','POPESTIMATE2015']

# Apply across the dataframe with a lambda
df.apply(lambda x:np.max(x[rows]), axis=1).head()

# A lambda is an unamed function in python. In this case, it takes a single parameter,x,
# and returns a single value.
```

```python
# To didvide the states into four categories
# Write a customized function that returns the region based on the state regions

def get_state_region(x):
    northeast= ['Connecticut', 'Maine', 'New Hampshire', 'Rhode Island', 'Vermont']
    midwest= ['Illinois', 'Indiana', 'Michigan', 'Ohio', 'Wisconson', 'Iowa']
    south= ['Delware', 'Florida', 'Georgia', 'Maryland', 'North Carolina', 'Alabama']
    west= ['Arizona', 'Colorado', 'Idaho', 'Montana', 'Nevada', 'New Mexico']
    
    if x in northeast:
        return 'northeast'
    elif x in midwest:
        return 'midwest'
    elif x in south:
        return 'south'
    else:
        return 'west'
```

```python
# To create a new column called Region, which shows the state's
# region, we can use the customized function and the apply function.
# The customized function works on the state name column, so we will
# set the apply function on the state name column and pass the customized
# function into the apply function
df['state_region']= df['STNAME'].apply(lambda x: get_state_region(x))
```
